```python
#Basic Data Science Libraries
import pandas as pd
import numpy as np

#Visualization libraries

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# For File Upload
from google.colab import files

#To set nice styles
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load Titanic dataset (great for exploration)
df = sns.load_dataset('titanic')

# First look
print("Shape:", df.shape)
df.head(10)
```

```python
# Basic info
df.info()
```

```python
# Summary statistics
df.describe(include='all')
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Fill missing age with median (good practice)
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
# Fill missing embarked with most common
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
```

```python
# Drop deck column (too many missing)
df.drop('deck', axis=1, inplace=True)
```

```python
# Fill missing embark_town with most common
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
```

```python
# Check again
df.isnull().sum()
```

```python
px.bar(df, x='sex', color='survived', barmode='group',
       title="Survival Count by Gender",
       labels={'sex': 'Gender', 'survived': 'Survived (1=Yes)'})
```

```python
px.histogram(df, x='age', color='survived', nbins=30,
             title="Age Distribution by Survival",
             labels={'age': 'Age'})
```

```python
px.scatter(df, x='age', y='fare', color='survived', size='pclass',
           hover_data=['sex', 'embarked'],
           title="Fare vs Age Colored by Survival")
```
